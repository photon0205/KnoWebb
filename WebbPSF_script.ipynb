```python
import matplotlib as plt
import numpy as np
import pandas as pd
import webbpsf 
import synphot
from stsynphot import grid_to_spec
import poppy

```

```python
a=webbpsf.NIRCam()
a.filter = 'F444W'
src1 = {'wavelengths': [3.8e-6, 4.4e-6, 4.9e-6], 'weights': [0.0, 1, 0.0]}
psf= a.calc_psf(source= src1,fov_arcsec=10)
webbpsf.display_psf(psf,title='Ideal Image')
plt.pyplot.savefig(f'images/NIRCam_Ideal.jpeg',dpi=150)
nc= webbpsf.NIRCam()
nc.filter = 'F444W'
for i in range(1,11):
    src = {'wavelengths': [3.8e-6, 4.4e-6, 4.9e-6], 'weights': [0.5-i*0.05, 0.1*i, 0.5-0.05*i]}
    psf10= nc.calc_psf(source = src,fov_arcsec=10)
    fig, (ax_1, ax_2) = plt.pyplot.subplots(1, 2,figsize=(10, 3.5))
    webbpsf.display_psf(psf10,ax=ax_1)
    webbpsf.display_psf_difference(psf,psf10,title='Difference',cmap='RdBu_r',ax=ax_2)
    plt.pyplot.savefig(f'images/NIRCam${i}.jpeg',dpi=150)
```
